Necesitamos cambiar el nombre de las imágenes para añadir la info de la lesión en el nombre. Es más cómodo y además nos permite separar por mass y calcification, que es necesario para la segmentación y más cosas.
Para ello, usaremos los csv que traen la info de cada imagen de INbreast:

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
INbreast = pd.read_csv('INbreastExcel.csv', sep = ';')
INbreast.info()
```

```python

# Normalizar columnas de lesión
def get_lesion_type(row):
    tipos = []

    # OJO: asegúrate de que los nombres de columnas están normalizados
    if pd.notna(row["Mass "]):        # o "Mass" si ya lo limpiaste
        tipos.append("MASS")
    if pd.notna(row["Micros"]):
        tipos.append("MICRO")
    if pd.notna(row["Distortion"]):
        tipos.append("DISTORTION")
    if pd.notna(row["Asymmetry"]):
        tipos.append("ASYMMETRY")

    if len(tipos) == 0:
        return "NONE"

    return "_".join(tipos)

INbreast["LesionType"] = INbreast.apply(get_lesion_type, axis=1)  #crea una nueva columna con los nombres 
print(INbreast.columns.tolist())
```

```python
INbreast[pd.isna(INbreast["File Name"])]
```

```python
#Eliminamos las filas con NaN:
INbreast = INbreast[INbreast["File Name"].notna()].copy()
```

```python
#Los nombres actuales empiezan con el filename. Así, solo tenemos que añadir los datos a la foto que 
#coincida con el filename
#Se añadirá el tipo de lesión y el bi-rads, que es como una clasificación subjetiva de la gravedad
import os
import pandas as pd

# === CONFIGURACIÓN ===
images_dir = "Imagenes mascaras segmentacion INbreast"  # carpeta donde están las imágenes originales
output_dir = "Imagenes mascaras segmentacion INbreast RENOMBRADAS"  
# carpeta donde guardarás las imágenes renombradas

def find_image_by_prefix(prefix, folder):
    """
    Busca el primer archivo en 'folder' cuyo nombre empiece por 'prefix'.
    Devuelve la ruta completa o None si no existe.
    """
    prefix = str(prefix)

    for fname in os.listdir(folder):
        if fname.startswith(prefix):
            return fname

    return None

os.makedirs(output_dir, exist_ok=True)



# === RENOMBRAR ===
for idx, row in INbreast.iterrows():

    # Nombre original del archivo (INbreast usa números tipo 205881)
    filename_dataset = str(int(row["File Name"]))
    original_name = find_image_by_prefix(filename_dataset, images_dir)

    try:
        original_path = os.path.join(images_dir, original_name)
    except:
        print(f"NO ENCONTRADO: {filename_dataset}")
        continue

    # Construir nuevo nombre
    lesion = row["LesionType"]
    birads = row['Bi-Rads']

    new_name = f"{lesion}_bi-rads_{birads}_{original_name}"
    new_path = os.path.join(output_dir, new_name)

    # copiar:
    import shutil
    original_path = os.path.join(images_dir, original_name)
   
    shutil.copy(original_path, new_path)
    print(f'imagen {original_name} renombrada', '\r')

print("Renombrado completado.")
```
